### Reading tables from pdf

```python
def extract_table(pdf):
    with open('foo.pdf', 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text

pdf_text = extract_table('foo.pdf')
```

```python
import pdfplumber

# Open the PDF file
with pdfplumber.open('foo.pdf') as pdf:
    # Extract tables from each page
    tables = []
    for page in pdf.pages:
        table = page.extract_tables()
        tables.extend(table)

# Print the extracted tables
for table in tables:
    print(table)
```

```python
def replace_newline(cell):
    if isinstance(cell, str):
        return cell.replace('\n', ' ')
    else:
        return cell
```

```python
df = pd.DataFrame(table)
df = df.applymap(replace_newline)
df
```
